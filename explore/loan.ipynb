```python
data = pd.read_csv("../data/hmeq_Loan_Default.csv")
data.head(15)
```

BAD → Did the customer default? (1 = Yes, 0 = No): is the customer owing? 

LOAN → Loan amount : The amount of money the customer wants to borrow. 

MORTDUE → Mortgage due: How much the person still owes on their mortgage (their house loan) 

VALUE → Property value: The estimated current value of their house. 

YOJ → Years on job: Why the customer is taking the loan.   Home improvement,           Debt consolidation 

DEROG → Number of major derogatory marks on a credit report.Serious negative events like:Bankruptcy   Very late payments Charge-offs 

DELINQ → Number of delinquencies: Number of times the customer was (late on payments). 

CLAGE → Age of oldest credit line(in months).CLAGE = 120 → the oldest account has been open 10 years. 

CLNO → Number of credit lines: Total number the person has ever opened. CLNO = 25 is normal. 5 is limited credit history. 

DEBTINC → Debt-to-income ratio: Shows how much of their income is already used to pay debts. DEBTINC = 36 → 36% of income is going to debt. 
Higher debt-to-income usually means higher risk of default.

# Check percentage of missing values
#### missing value rules
* Check percentage missing
  - < 5% → usually safe to drop or simple impute
  - 5–30% → impute carefully
  - > 30% → consider dropping the variable entirely

```python
# check missing values

data.isnull().sum()
```

```python
missing_data = data.isnull().sum() * 100 / len(data)
missing_percentage = missing_data.apply(lambda x: f"{x:.0f}% missing")
print(missing_percentage)
```

# Identify the type of missingness
  - Missing Completely At Random MCAR
  - Missing At Random MAR
  - Missing Not At Random MNAR

```python
# visualize missing values

plt.figure(figsize=(12,6))
sns.heatmap(data.isnull(), cbar=True, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```
